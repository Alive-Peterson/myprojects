### Recommendation System using Collaborative and Content based filtering

```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("ml-latest-small/movies.csv")
ratings = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
ratings.head()
```

```python
movies.head()
```

#### merging both datasets

```python
df = ratings.merge(movies, on='movieId')

df.head()
```

##### Checking nulls

```python
df.isnull().sum()
```

#### Collaborative Filtering

##### Create User Item Matrix

```python
user_movie_matrix=df.pivot_table(index='userId',columns='title', values='rating')
```

```python
user_movie_matrix.head()
```

##### Handling Missing Values

```python
user_movie_matrix = user_movie_matrix.fillna(0)
```

```python
user_movie_matrix.head()
```

##### Calcualting Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
movie_similarity = cosine_similarity(user_movie_matrix.T)
```

##### Convert it into DataFrame

```python
movie_similarity_df = pd.DataFrame(
    movie_similarity,
    index=user_movie_matrix.columns,
    columns=user_movie_matrix.columns
)
```

```python
movie_similarity_df.head()
```

##### Test Recommendation

```python
movie_name = 'Toy Story (1995)'
similar_movies = movie_similarity_df[movie_name].sort_values(ascending=False)[1:6]
print(similar_movies)
```

##### Personalized Recommendation

##### select user

```python
user_id = 1
```

##### get user ratings

```python
user_ratings = user_movie_matrix.loc[user_id]
user_ratings.head()
```

##### Filter watched movies by user

```python
watched_movies = user_ratings[user_ratings>0]
print("Movies watched by user:", len(watched_movies))
watched_movies.head()
```

##### Similarity score

```python
similarity_scores = pd.Series(dtype=float)
```

```python
for movie, rating in watched_movies.items():
    similar_movies = movie_similarity_df[movie]
    similarity_scores = similarity_scores.add(similar_movies * rating, fill_value=0)
```

##### Remove Already Watched Movies

```python
similarity_scores = similarity_scores.drop(watched_movies.index, errors='ignore')
```

##### Get Top Recommendations

```python
recommended_movies = similarity_scores.sort_values(ascending=False).head(10)
recommended_movies
```

#### Content Based Filtering

```python
movies[['title','genres']].head()
```

##### Convert Genres to usable format

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(tokenizer=lambda x: x.split('|'))

genre_matrix = movies['genres'].str.get_dummies(sep='|')
```

##### compute similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_similarity = cosine_similarity(genre_matrix)
```

##### Converting it into dataframe

```python
genre_similarity_df = pd.DataFrame(genre_similarity, index=movies['title'], columns=movies['title'])
```

##### Testing it on similar genre

```python
movie_name = "Toy Story (1995)"

genre_similarity_df[movie_name].sort_values(ascending=False)[1:6]
```

#### Hybrid Recommendation System (Collaborative + Content based filtering)

```python
def recommend_movies(movie_name, top_n=10):
    collab_scores = movie_similarity_df[movie_name]
    content_scores = genre_similarity_df[movie_name]
    
    final_scores = (collab_scores * 0.7) + (content_scores * 0.3)
    final_scores = final_scores.drop(movie_name)
    
    return final_scores.sort_values(ascending=False).head(top_n)

recommend_movies("Toy Story (1995)")
```

#### Making CSV for Dashboard

##### Average rating per movie

```python
movie_avg_rating = df.groupby('title')['rating'].mean().reset_index()
movie_avg_rating.columns = ['title', 'avg_rating']
```

##### Number of ratings (popularity)

```python
movie_count = df.groupby('title')['rating'].count().reset_index()
movie_count.columns = ['title', 'rating_count']
```

##### Merge everything

```python
movie_stats = movie_avg_rating.merge(movie_count, on='title')
movie_stats = movie_stats.merge(movies[['title', 'genres']], on='title')

movie_stats.head()
```

##### Export to CSV

```python
movie_stats.to_csv("movie_dashboard_data.csv", index=False)
```

### Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("OnlineRetail.csv", encoding='ISO-8859-1', low_memory=False)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

##### Handle Missing Values

```python
df.isnull().sum()
```

##### Removing null/blank values

```python
df = df.dropna(subset=['CustomerID'])
```

##### Removing negative or zero values for quantity and UnitPrice

```python
df = df[df['Quantity'] > 0]
```

```python
df = df[df['UnitPrice'] > 0]
```

##### Convert Date Column

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

##### Total Price

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df.info()
```

```python
df.describe()
```

##### exporting cleaned dataset

```python
df.to_csv("cleaned_data.csv", index=False)
```

#### Summary

- Removed missing CustomerID records
- Filtered out negative quantities and prices
- Converted InvoiceDate to datetime format
- Created TotalPrice column

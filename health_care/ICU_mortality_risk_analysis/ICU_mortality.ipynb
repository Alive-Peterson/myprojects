#### Checking Value Counts

```python
cols = [
    'gender',
    'admission_type',
    'ventilation_required',
    'vasopressor_used',
    'sepsis_flag',
    'mortality_label'
]

for col in cols:
    print(f"\n{'='*50}")
    print(f"Value Counts for {col}")
    print(df[col].value_counts(dropna=False))
```

#### Feature Engineering

##### Age Bucket

```python
df['age_bucket'] = pd.cut(
    df['age'],
    bins=[17,29,39,49,59,69,90],
    labels=['18-29','30-39','40-49','50-59','60-69','70+']
)
```

##### LOS Bucket

```python
df['los_bucket'] = pd.cut(
    df['length_of_stay_days'],
    bins=[0,3,7,14,30],
    labels=['1-3 Days','4-7 Days','8-14 Days','15+ Days']
)
```

##### APACHE Bucket

```python
df['apache_bucket'] = pd.cut(
    df['apache_score'],
    bins=[0,10,20,30,40],
    labels=['0-10','11-20','21-30','31-40']
)
```

##### SOFA Bucket

```python
df['sofa_bucket'] = pd.cut(
    df['sofa_score'],
    bins=[-1,5,10,15,20],
    labels=['0-5','6-10','11-15','16-20']
)
```

#### Analysis check

##### Comorbidity Bucket

```python
pd.crosstab(
    df['age_bucket'],
    df['mortality_label'],
    normalize='index'
)*100
```

##### Mortality by ventilation

```python
pd.crosstab(
    df['ventilation_required'],
    df['mortality_label'],
    normalize='index'
)*100
```

##### Mortality by Sepsis

```python
pd.crosstab(
    df['sepsis_flag'],
    df['mortality_label'],
    normalize='index'
)*100
```

##### Mortality by Admission Type

```python
pd.crosstab(
    df['admission_type'],
    df['mortality_label'],
    normalize='index'
)*100
```

##### Mortality by APACHE Bucket

```python
pd.crosstab(
    df['apache_bucket'],
    df['mortality_label'],
    normalize='index'
)*100
```

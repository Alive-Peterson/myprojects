### Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("diabetic_data.csv")
df.head()
```

```python
(df == "?").sum().sort_values(ascending=False)
```

```python
# Replacing "?" with NaN
df.replace("?", np.nan, inplace=True)
```

```python
df.head(15)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# Dropping weight column since it has too many null values
df.drop(columns=["weight"], inplace=True)
```

```python
# fixing gender column
df["gender"] = df["gender"].replace("Unknown/Invalid", "Unknown")
```

##### Replacing missing values 

```python
text_cols = [
    "race",
    "payer_code",
    "medical_specialty",
    "diag_1",
    "diag_2",
    "diag_3",
    "max_glu_serum",
    "A1Cresult"
]

df[text_cols] = df[text_cols].fillna("Unknown")
```

```python
# verifying if nulls are there
df.isnull().sum()
```

##### Readmission flag

```python
df["ReadmissionFlag"] = (df["readmitted"] == "<30").astype(int)
```

##### Readmission category

```python
df["ReadmissionCategory"] = df["readmitted"].replace({
    "<30":"Within 30 days",
    ">30":"After 30 days",
    "NO":"Not readmitted"
})
```

##### Length of Stay Buckets

```python
df["LOS_Bucket"] = pd.cut(df["time_in_hospital"], bins=[0, 3, 7, 14, 100], labels=["1-3 Days","4-7 Days","8-14 Days","15+ Days"])
```

##### Total Prior Visits

```python
df["TotalPriorVisits"] = (df["number_outpatient"] + df["number_emergency"] + df["number_inpatient"])
```

##### Prior Visit Buckets

```python
df["PriorVisitBucket"] = pd.cut(df["TotalPriorVisits"], bins=[-1, 0, 2, 5, 100], labels=["None","1-2 Visits","3-5 Visits","5+ Visits"])
```

##### Medication Load

```python
df["MedicationLoad"] = pd.cut(df["num_medications"],bins=[0, 10, 20, 50, 100],
    labels=["Low","Medium","High","Very High"])
```

##### Admission type map

```python
admission_type_map = {
    1: "Emergency",
    2: "Urgent",
    3: "Elective",
    4: "Newborn",
    5: "Not Available",
    6: "Unknown",
    7: "Trauma Center",
    8: "Other"
}

df["AdmissionType"] = df["admission_type_id"].map(admission_type_map)
```

##### admission source map

```python
admission_source_map = {
    1: "Physician Referral",
    2: "Clinic Referral",
    3: "HMO Referral",
    4: "Transfer from Hospital",
    5: "Transfer from Skilled Nursing Facility",
    6: "Transfer from Health Care Facility",
    7: "Emergency Room",
    8: "Court/Law Enforcement",
    9: "Not Available",
    10: "Transfer from Critical Access Hospital",
    11: "Normal Delivery",
    13: "Sick Baby",
    14: "Extramural Birth",
    17: "Unknown",
    20: "Transfer from Another Facility",
    22: "Born Inside Hospital",
    25: "Transfer from Ambulatory Surgery Center"
}

df["AdmissionSource"] = df["admission_source_id"].map(admission_source_map)
```

##### discharge group

```python
def discharge_group(x):
    if x == 1:
        return "Home"

    elif x in [2,3,4,5]:
        return "Transferred"

    elif x in [6,7,8]:
        return "Home Health Care"

    elif x in [11,12,13,14,23,24]:
        return "Hospice"

    elif x in [18,25,26]:
        return "Psychiatric/Rehab"

    elif x in [22]:
        return "Expired"

    else:
        return "Other"

df["DischargeGroup"] = df["discharge_disposition_id"].apply(discharge_group)
```

##### Age sort

```python
age_sort = {
    "[0-10)":1,
    "[10-20)":2,
    "[20-30)":3,
    "[30-40)":4,
    "[40-50)":5,
    "[50-60)":6,
    "[60-70)":7,
    "[70-80)":8,
    "[80-90)":9,
    "[90-100)":10
}

df["AgeSort"] = df["age"].map(age_sort)
```

##### Exporting 

```python
df.to_csv(
    "hospital_readmission_final.csv",
    index=False
)
```
